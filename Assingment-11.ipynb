```python
import multiprocessing
l = [6,12,3,9]
new_l = []
def process_module(l):
    for i in l :
        new_l.append(i**3)
        
if __name__ == '__main__':
    process = multiprocessing.Process(target=process_module(l,))
    process.start()
    process.join()
print(new_l)    
```

```python
import multiprocessing

def num_1():
    print('1')
    
def num_2():
    print('2')

def num_3():
    print('3')
    
def num_4():
    print('4')
    
if __name__ == '__main__':
    process1 = multiprocessing.Process(target = num_1)
    process2 = multiprocessing.Process(target = num_2)
    process3 = multiprocessing.Process(target = num_3)
    process4 = multiprocessing.Process(target = num_4)
    
    process1.start()
    process2.start()
    process3.start()
    process4.start()
    process1.join()
    process2.join()
    process3.join()
    process4.join()
```
