```python
""" Scatter plots are the graphs that present the relationship between two variables in a data-set. It represents data points on a
    two-dimensional plane or on a Cartesian system. The independent variable or attribute is plotted on the X-axis, while the dependent
    variable is plotted on the Y-axis. These plots are often called scatter graphs or scatter diagrams.scatter plots the data in form of 
    point or we can change it using its function of matplotlib."""

import numpy as np
import matplotlib.pyplot as plt
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
scatter_graph = plt.scatter(x,y)
plt.xlabel("this is x axis")
plt.ylabel("this is y axis")
plt.title("this is graph for assignment-23")
scatter_graph   
```

```python
""" The subplots() function in the Pyplot module of the Matplotlib library is used to create a figure and a set of subplots."""

import numpy as np
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 100, 200, 300, 400, 500])
plt.subplot(1, 4, 1)
plt.plot(x,y)

x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([50, 20, 40, 20, 60, 70])
plt.subplot(1,4,2)
plt.plot(x,y)

x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([10, 20, 30, 40, 50, 60])
plt.subplot(1,4,3)
plt.plot(x,y)

x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([200, 350, 250, 550, 450, 150])
plt.subplot(1,4,4)
plt.plot(x,y)
```

```python
""" A bar plot or bar chart is a graph that represents the category of data with rectangular bars with lengths and heights 
    that is proportional to the values which they represent. The bar plots can be plotted horizontally or vertically. A bar 
    chart describes the comparisons between the discrete categorie """

import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.bar(company,profit)
```

```python
""" boxplot is a method for graphically demonstrating the locality, spread and skewness groups of numerical data through their quartiles
     It is also termed as box and whisker plot.A box plot is a chart that shows data from a five-number summary including one of the 
     measures of central tendency."""

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

box3 = [box1,box2]
plt.show(plt.boxplot(box3))
```
